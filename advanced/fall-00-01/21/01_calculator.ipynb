# Calculator


```python
from tkinter import *
from functools import partial



root = Tk()
root.title("Calculator")
# root.geometry("320x400")


operators = ("+", "-", "/", "*", "C")

# 
first_number, second_number = 0, 0
opt = None


def on_click(btn_text):
    global entry, first_number, opt
    
    entry.config(state=NORMAL)
    text = entry.get()

    if btn_text == "C":
        first_number, second_number, opt = 0, 0, None
        entry.delete(0, END)
        return
    
    elif btn_text == "=":
        second_number = float(text)
        entry.delete(0, END)

        res = 0
        if opt == "+":
            res = first_number + second_number
        elif opt == "-":
            res = first_number - second_number
        elif opt == "*":
            res = first_number * second_number
        elif opt == "/":
            try:
                res = first_number / second_number
            except ZeroDivisionError:
                res = "اشتباه زدی برادر"

        entry.insert(END, res)
        entry.config(state=DISABLED)
        return
    # 
    elif btn_text in operators:
        for operator in operators:
            if operator in text:
                entry.delete(0, END)
                break
        else:
            first_number = float(text) if text else 0
            opt = btn_text
            entry.delete(0, END)
    else:
        # +1
        # 
        for operator in operators:
            if operator in text:
                entry.delete(0, END)  

    # 
    entry.insert(END, btn_text)
    entry.config(state=DISABLED)


#entry = Entry(root, width=25, font="Hevelica 14", relief=SUNKEN, borderwidth=2)
entry = Entry(root, width=30, font=("Hevelica", 14), relief=SUNKEN, borderwidth=2, state=DISABLED)
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10, ipadx=10, ipady=5)


"""
1 2 3 +
4 5 6 -
7 8 9 *
C 0 = /
"""

btn_collections = [
    ("1", "2", "3", "+"),
    ("4", "5", "6", "-"),
    ("7", "8", "9", "*"),
    ("C", "0", "=", "/"),
]

for row, btns in enumerate(btn_collections):
    for col, btn in enumerate(btns):
        button = Button(
            root, 
            text=btn, 
            width= 10, 
            height= 3, 
            command= partial(on_click, btn), 
            relief=RAISED, 
            bd=2,
            font=("Hevelica", 8)
        )
        button.grid(row=row + 1, column=col)

root.mainloop()
```
