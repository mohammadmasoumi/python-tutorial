## Function VS Method

### Function

Call below expression a **function** whenever it's in file(not in classes).

```python
def function(a, b):
    return a + b
```

### Method

Call below expression a **method** whenever it's behaviour of an **object**.

```python

class Student:

    def move(self):
        pass

```


## Argument VS Params


**params:** function input parameters

**args:** input when calling function


```python
def add(a, b): # params
    # assign args to params
    # a = 10
    # b = 12
    return a + b

add(10, 12) # args
```


# Invoke/Call

```python

def add(a, b):
    return a + b

# invoke
# call
add(10, 12)
add(12, 13)

```
