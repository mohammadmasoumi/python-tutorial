- get 5 scores from input
- find maximum 
- calc difference of maximum and 20
- add difference to all scores
- create new list

```python
# 1 - get all scores from input
# Check all scores should be between 0 and 20

#         20, 13, 
scores = []
courses = ["math", "science", "art", "pe", "literature"]


# courses[0] -> scores[0]
# "math" -> 20


math_score = float(input())

# (0 <= math_score <= 20)
# True => not True => False
# if math_score < 0 and 20 < math_score:    
if not (0 <= math_score <= 20):
    print("invalid score!")
    # exit()
else:
    scores.append(math_score)


science_score = float(input())
if not (0 <= science_score <= 20):
    print("invalid score!")
    # exit()
else:
    scores.append(science_score)
    

art_score = float(input())
if not (0 <= art_score <= 20):
    print("invalid score!")
    # exit()
else:
    scores.append(art_score)


pe_score = float(input())
if not (0 <= pe_score <= 20):
    print("invalid score!")
    # exit()
else:
    scores.append(pe_score)


literature_score = float(input())
if not (0 <= literature_score <= 20):
    print("invalid score!")
    # exit()
else:
    scores.append(literature_score)

print(f"{courses[0]: <10}: {scores[0]}")
print(f"{courses[1]: <10}: {scores[1]}")
print(f"{courses[2]: <10}: {scores[2]}")
print(f"{courses[3]: <10}: {scores[3]}")
print(f"{courses[4]: <10}: {scores[4]}")
print("--------------------")


# 2
# maximum = max(math_score, science_score, art_score, pe_score, literature_score)
maximum = max(scores)
print(f"maximum : {maximum}")


# 3
difference = 20 - maximum
print(f"diff: {difference}")

# 4
# math_score += diff
# science_score += diff
# art_score += diff
# pe_score += diff
# literature_score += diff

scores[0] += difference # math
scores[1] += difference # scient
scores[2] += difference # art
scores[3] += difference # pe
scores[4] += difference # literature
print("--------------------")

# 5
print(f"{courses[0]: <10}: {scores[0]}")
print(f"{courses[1]: <10}: {scores[1]}")
print(f"{courses[2]: <10}: {scores[2]}")
print(f"{courses[3]: <10}: {scores[3]}")
print(f"{courses[4]: <10}: {scores[4]}")
```
