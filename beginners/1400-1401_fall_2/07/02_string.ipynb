# Oneline input



### Getting inputs from user in one line


Instead of getting inputs in 4 lines
```python
12
13
14
15
```
we can get inputs in oneline and split them with space

```python
12 13 14 15

input().split()
```





````python
# Oneline input

# 12 ENTER
# 13 ENTER
# 14 ENTER
# 15 ENTER

#  0  1  2  3
# 12 13 14 15 ENTER

s = input()

# 12 13 14 15
# 12, 13, 14, 15
# print(12, 13, 14, 15, sep=", ")
print(s, sep=", ")
print(s.split(), sep=", ")
````
