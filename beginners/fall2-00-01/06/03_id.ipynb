## Immutable - تغییر ناپذیر

If you want to change an immutable obj you should re-assign it to the variable.

- str
- int
- bool
- float
- tuple

## Mutable

دیتا تایپ هایی هستند که می توانید درجا تغییر دهید.


- list
- set
- dict


```python
# id - built-in

str1 = "Hello world"
# unique identifier
print(id(str1))

# -5 , 255 -> cache
a = 10
b = 10

print("a:", id(a))
print("b:", id(b))


c = 101112
d = 101
112

print("c:", id(c))
print("d:", id(d))
```

```python
a = [12, 13]

print(id(a))
print(id(a[0]))
print(id(a[1]))
print(id(a[1]) - id(a[0]))
print("-----------------------")

# append
a.append(14)
print(id(a[1]) - id(a[0]))
print(id(a[2]) - id(a[1]))
print("-----------------------")

# insert
a.insert(1, 15)
print(id(a[1]) - id(a[0]))
print(id(a[2]) - id(a[1]))
print(id(a[3]) - id(a[2]))
print("-----------------------")

# pop
a.pop()
print(id(a[1]) - id(a[0]))
print(id(a[2]) - id(a[1]))
print("-----------------------")

# remove
a.remove(15)
print(id(a[1]) - id(a[0]))
print("-----------------------")

```

```python
a = 12
b = 12

# -5 - 255
print(id(a))
print(id(b))

# check the value of a and b
# True or False
print(a == b)

# id(a) == id(b)
print(a is b)

print("------------------")
c = 1000
d = 1000

print(id(c))
print(id(d))

print(c == d) # check the value of c and d

# id(c) == id(d)
print(c is d) # check the reference c and d
print("------------------")

# str
# the same container
e = "hello"
f = "hello"

e1 = "helloooooooooooooooooooooo"
f1 = "helloooooooooooooooooooooo"

print(id(e))
print(id(f))
print(id(e1))
print(id(f1))

print(e == f)
print(e is f)
print("------------------")

# float
a1 = 12.2
b1 = 12.2

c1 = 1000.1
d1 = 1000.1

print(id(a1))
print(id(b1))
print(id(c1))
print(id(d1))

print(a1 is b1)
print(c1 is d1)
```
